## Prepare data for candlestick chart

## Create the candlestick chart

### Subtask:
Use the extracted data to generate a candlestick chart using `plotly.graph_objects`.


**Reasoning**:
Generate a candlestick chart using plotly.graph_objects.



```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(x=df_ohlc.index,
                open=df_ohlc['open'],
                high=df_ohlc['high'],
                low=df_ohlc['low'],
                close=df_ohlc['close'])])

fig.update_layout(
    title='AAPL Candlestick Chart (2023)',
    xaxis_title='Date',
    yaxis_title='Price'
)

fig.show()
```

## Prepare data for candlestick chart with volume

### Subtask:
Extract the necessary columns ('open', 'high', 'low', 'close', 'volume' and the time index) from the DataFrame `df`.


**Reasoning**:
Create a new DataFrame with the relevant columns and rename them for the candlestick chart, then display the head.



```python
df_ohlcv = df[['o', 'h', 'l', 'c', 'v']]
df_ohlcv.columns = ['open', 'high', 'low', 'close', 'volume']
display(df_ohlcv.head())
```

## Create the candlestick chart with volume

### Subtask:
Use the extracted data to generate a candlestick chart with a volume subplot using `plotly.graph_objects`.


**Reasoning**:
Generate a candlestick chart with a volume subplot using plotly.graph_objects.



```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.03,
                    subplot_titles=('Candlestick Chart', 'Volume'))

fig.add_trace(go.Candlestick(x=df_ohlcv.index,
                             open=df_ohlcv['open'],
                             high=df_ohlcv['high'],
                             low=df_ohlcv['low'],
                             close=df_ohlcv['close'],
                             name='Price'), row=1, col=1)

fig.add_trace(go.Bar(x=df_ohlcv.index, y=df_ohlcv['volume'], name='Volume'), row=2, col=1)

fig.update_layout(
    title='AAPL Candlestick Chart with Volume (2023)',
    xaxis_rangeslider_visible=False
)

fig.show()
```

## Display the chart

### Subtask:
Display the generated candlestick chart with volume.


**Reasoning**:
Display the generated candlestick chart with volume.



```python
fig.show()
```
